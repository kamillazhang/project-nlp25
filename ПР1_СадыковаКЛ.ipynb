```python
# title АНАЛИЗАТОР ТЕКСТОВОЙ СТАТИСТИКИ
print("АНАЛИЗАТОР ТЕКСТОВОЙ СТАТИСТИКИ")
print(" ")

text = input("Введите текст для анализа: ")

chars = len(text)
words = len(text.split())
sentences = text.count('.') + text.count('!') + text.count('?')
print(" ")
print(f"=== СТАТИСТИКА ТЕКСТА ===")
print(f"Общее количество символов: {chars}")
print(f"Количество слов: {words}") 
print(f"Количество предложений: {sentences if sentences > 0 else 1}")




# title КАЛЬКУЛЯТОР КАЛОРИЙ ДЛЯ ТРЕНИРОВКИ
print("КАЛЬКУЛЯТОР КАЛОРИЙ ДЛЯ ТРЕНИРОВКИ")
print(" ")

run = int(input("Введите минуты бега: "))
pushup = int(input("Введите минуты отжиманий: "))
squat = int (input("Введите минуты приседаний: "))


run_calories = run * 10 
pushup_calories = pushup * 8
squat_calories = squat * 6
total_calories = run_calories + pushsup_calories + squat_calories

print(" ")
print(f"=== РЕЗУЛЬТАТЫ ТРЕНИРОВКИ ===")
print(f"Бег: {run} мин × 10 ккал/мин = {run_calories} ккал")
print(f"Отжимания: {pushup} мин × 8 ккал/мин = {pushup_calories} ккал")
print(f"Приседания: {squat} мин × 6 ккал/мин = {squat_calories} ккал")
print("------------------------------")
print(f"ОБЩЕЕ: {total_calories} ккал")





import getpass
# title СОЗДАНИЕ ПЛЕЙЛИСТА ЛЮБИМЫХ ПЕСЕН
print("СОЗДАНИЕ ПЛЕЙЛИСТА ЛЮБИМЫХ ПЕСЕН")
print("")
print("Добавьте 5 любимых песен:")

playlist = []

for song_number in range(5):
    song = input(f"Песня {song_number+1}: ")
    playlist.append(song)
print("")
print("\n=== ВАШ ПЛЕЙЛИСТ ===")
for song_number, song in enumerate(playlist, 1):
    print(f"{song_number}. {song}")
print()
print(f"Первая песня: {playlist[0]}")
print(f"Последняя песня: {playlist[-1]}")
print(f"Всего песен: {len(playlist)}")
print("")
try:
    # Ввод без отображения
    song_to_remove = int(getpass.getpass(""))
    
    if 1 <= song_to_remove <= len(playlist):
        removed_song = playlist.pop(song_to_remove - 1)
        print(f"Удаляем песню №{song_to_remove}: {removed_song}")
    else:
        print("Неверный номер песни!")
except ValueError:
    print("Пожалуйста, введите число!")
print("")
print("\n=== ОБНОВЛЕННЫЙ ПЛЕЙЛИСТ ===")
for song_number, song in enumerate(playlist, 1):
    print(f"{song_number}. {song}")





# title ПОИСК ОБЩИХ ДРУЗЕЙ В СОЦСЕТЯХ
print("ПОИСК ОБЩИХ ДРУЗЕЙ В СОЦСЕТЯХ")
print("")
friends1_input = input("Введите друзей пользователя 1 (через запятую):\n")
friends2_input = input("\nВведите друзей пользователя 2 (через запятую):\n")

friends1 = set(name.strip() for name in friends1_input.split(','))
friends2 = set(name.strip() for name in friends2_input.split(','))

common_friends = friends1 & friends2

unique_friends1 = friends1 - friends2
unique_friends2 = friends2 - friends1

print("")
print("\n=== РЕЗУЛЬТАТЫ АНАЛИЗА ===")
print(f"Друзья пользователя 1: {len(friends1)} человек")
print(f"Друзья пользователя 2: {len(friends2)} человек")

if common_friends:
    print(f"Общие друзья: {', '.join(sorted(common_friends))}")
else:
    print("Общие друзья: нет общих друзей")

print(f"Количество общих друзей: {len(common_friends)}")
print("")
if unique_friends1:
    print(f"Уникальные друзья user1: {', '.join(sorted(unique_friends1))}")
else:
    print("Уникальные друзья user1: нет уникальных друзей")

if unique_friends2:
    print(f"Уникальные друзья user2: {', '.join(sorted(unique_friends2))}")
else:
    print("Уникальные друзья user2: нет уникальных друзей")




import math

# title РАСЧЕТ РАССТОЯНИЯ МЕЖДУ ТОЧКАМИ
print("РАСЧЕТ РАССТОЯНИЯ МЕЖДУ ТОЧКАМИ")
print("")
print("Введите координаты точки A:")
lat1 = float(input("Широта: "))
lon1 = float(input("Долгота: "))
print("")
print("Введите координаты точки B:")
lat2 = float(input("Широта: "))
lon2 = float(input("Долгота: "))

point_a = (lat1, lon1)
point_b = (lat2, lon2)

distance = math.sqrt((lat2 - lat1)**2 + (lon2 - lon1)**2)

print("\n=== РЕЗУЛЬТАТЫ ===")
print(f"Точка A: {point_a}")
print(f"Точка B: {point_b}")
print(f"Расстояние между точками: {distance:.4f}")




# title МОЯ КУЛИНАРНАЯ КНИГА
print("МОЯ КУЛИНАРНАЯ КНИГА")
print("")

recipes = {
    "Жареный картофель с грибами": {
        "ingredients": ["картофель", "грибы", "лук", "масло растительное", "соль"],
        "time": 40
    },
    "Сладкие тыквенные драники": {
        "ingredients": ["тыква", "яйца", "мука", "соль", "сахар", "масло растительное"],
        "time": 15
    },
    "Омлет": {
        "ingredients": ["яйца", "масло подсолнечное", "соль", "перец"],
        "time": 10
    }
}

print("=== ВСЕ РЕЦЕПТЫ ===")
for i, recipe_name in enumerate(recipes.keys(), 1):
    print(f"{i}. {recipe_name}")
print("")

print("=== ДЕТАЛИ РЕЦЕПТОВ ===")
for recipe_name, recipe_info in recipes.items():
    print(f"{recipe_name}:\n• Ингредиенты: {', '.join(recipe_info['ingredients'])}\n• Время: {recipe_info['time']} мин")
    print("")

fastest_recipe = min(recipes.items(), key=lambda x: x[1]['time'])
slowest_recipe = max(recipes.items(), key=lambda x: x[1]['time'])

print(f"Самый быстрый: {fastest_recipe[0]} ({fastest_recipe[1]['time']} мин)")
print(f"Самый долгий: {slowest_recipe[0]} ({slowest_recipe[1]['time']} мин)")




# title ОПРЕДЕЛИТЕЛЬ ВРЕМЕНИ СУТОК
print("ОПРЕДЕЛИТЕЛЬ ВРЕМЕНИ СУТОК")
print("")
hour = int(input("Введите текущий час (0-23): "))
print("")
if hour < 0 or hour > 23:
    print("Ошибка! Введите число от 0 до 23")
else:
    if 5 <= hour < 12:
        time_of_day = "утро"
        recommendation = "Самое время для зарядки и полезного завтрака!"
    elif 12 <= hour < 17:
        time_of_day = "день"
        recommendation = "Отличное время для продуктивной работы и учебы!"
    elif 17 <= hour < 22:
        time_of_day = "вечер"
        recommendation = "Пора отдыхать и проводить время с семьей!"
    else:
        time_of_day = "ночь"
        recommendation = "Время для сна!"

    print(f"Сейчас {time_of_day}!")
    print(f"{recommendation}")




# title ТАБЛИЦА УМНОЖЕНИЯ
print("ТАБЛИЦА УМНОЖЕНИЯ")
print("")
number = int(input("Введите число для таблицы умножения: "))
print("")
print(f"\nТаблица умножения для числа {number}:")
for i in range(1, 11):
    result = number * i
    print(f"{number} × {i} = {result}")





# title УМНЫЙ КАЛЬКУЛЯТОР ОПЕРАЦИЙ С ТЕКСТОМ
print("УМНЫЙ КАЛЬКУЛЯТОР ОПЕРАЦИЙ С ТЕКСТОМ")
print("")
print("Доступные команды:")
print("- uppercase: преобразовать в верхний регистр")
print("- lowercase: преобразовать в нижний регистр")
print("- length: посчитать длину текста")
print("- words: посчитать количество слов")
print("- reverse: перевернуть текст")
print()
print("")
command = input("Введите команду: ").strip().lower()
text = input("Введите текст: ")

match command:
    case "uppercase":
        result = text.upper()
        print(f"Результат: {result}")
    
    case "lowercase":
        result = text.lower()
        print(f"Результат: {result}")
    
    case "length":
        length = len(text)
        print(f"Результат: Длина текста: {length} символов")
    
    case "words":
        words_count = len(text.split())
        print(f"Результат: Количество слов: {words_count}")
    case "reverse":
        reversed_text = text[::-1]
        print(f"Результат: {reversed_text}")
    
    case _:
        print("Ошибка: Неизвестная команда!")
        print("Доступные команды: uppercase, lowercase, length, words, reverse")


```
